```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
file_path = "cleaned_data.csv"
df = pd.read_csv(file_path)

# Basic info
print("Shape of dataset:", df.shape)
print("\nColumn Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())

df.head()
```

```python
# Statistical summary
print("Descriptive Statistics:\n")
display(df.describe(include='all'))

# Check for unique values in each column
print("\nUnique value counts:")
print(df.nunique())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.hist(figsize=(12,8), bins=25, edgecolor='black')
plt.suptitle("Distribution of Numeric Features", fontsize=14)
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top Categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    plt.figure(figsize=(8,3))
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

```python
# Save summary report to CSV
eda_summary = df.describe(include='all').transpose()
eda_summary.to_csv("eda_summary.csv")

print("✅ EDA Complete! Summary saved as 'eda_summary.csv'")
```
